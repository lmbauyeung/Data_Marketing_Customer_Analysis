```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Activity 3

```python
ca_df = pd.read_csv('Data_Marketing_Customer_Analysis_Round2.csv')
```

```python
ca_df
```

```python
ca_df.info()
```

```python
ca_df.drop(columns=['Unnamed: 0'], inplace = True)
```

```python
ca_df.info()
```

```python
Effective_month_list = []
for i in ca_df['Effective To Date']:
    b = i.partition('/')[0]
    Effective_month_list.append(int(b))
np.unique(Effective_month_list)
```

```python
ca_df['Effective Month'] = Effective_month_list
```

```python
first_quarter_filter = ca_df['Effective Month'] <=3
```

```python
ca_df[first_quarter_filter]
```

```python
# total number of responses
total_responses = ca_df['Response'].value_counts()
total_responses
```

```python
fg, ax = plt.subplots(1,1, figsize=(7,5))
ax.set_title('Total Number of Responses', size = 18)
ax.bar(total_responses.index, total_responses, color = ('red', 'green'))
```

```python
# response rate by sales channel
sales_channel = ca_df.groupby(['Sales Channel','Response'])['Customer'].count().reset_index()
sales_channel
```

```python
sns.set_palette('RdYlGn')
palette_response = {'Yes' : 'C5', 'No' : 'C0'}
fg2, ax2 = plt.subplots(figsize = (10,5))
ax2.set_title('Response Rate by Sales Channel', size = '18')
sns.barplot(x = sales_channel['Sales Channel'], y = sales_channel['Customer'], hue = sales_channel['Response'], ax = ax2, palette = palette_response)
```

```python
# response rate by total claim amount
plt.figure(figsize = (10,5))
sns.boxplot(x = 'Total Claim Amount', y = 'Response', data = ca_df)
plt.title('Response Rate by Total Claim Amount', size = '18')
```

```python
# response rate by income
plt.figure(figsize = (10,5))
sns.boxplot(x = 'Income', y = 'Response', data = ca_df)
plt.title('Response Rate by Income', size = '18')
```

```python
# correlation heatmap
corr = round(ca_df.corr(),3)
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style('white'):
    f3, ax3 = plt.subplots(figsize = (10,5))
    ax3 = sns.heatmap(corr, mask = mask, cmap = 'coolwarm', vmin = -1, vmax = 1, annot = True)
```
