# Activity 1

```python
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
```

```python
def rename_column_file1(file1):
    file1.rename(columns={'Customer':'customer_id', 'ST':'state', 'GENDER':'gender', 'Education':'education', 'Customer Lifetime Value':'clv',
       'Income':'income', 'Monthly Premium Auto':'monthly_premium_auto', 'Number of Open Complaints':'number_of_open_complaints',
       'Policy Type':'policy_type', 'Vehicle Class':'vehicle_class', 'Total Claim Amount':'total_claim_amount'}, inplace=True)
    return file1
```

```python
rename_column_file1(file1)
```

```python
rename_column_file1(file2)
```

```python
def rename_column_file3(file3):
    file3.rename(columns={'Customer':'customer_id', 'State':'state', 'Gender':'gender', 'Education':'education', 'Customer Lifetime Value':'clv',
       'Income':'income', 'Monthly Premium Auto':'monthly_premium_auto', 'Number of Open Complaints':'number_of_open_complaints',
       'Policy Type':'policy_type', 'Vehicle Class':'vehicle_class', 'Total Claim Amount':'total_claim_amount'}, inplace=True)
    return file3
```

```python
rename_column_file3(file3)
```

```python
ca_df = pd.concat([file1,file2,file3], axis=0).reset_index(drop=True)
ca_df
```

```python
# deleting the column customer
def drop_column(ca_df):
    ca_df.drop(columns=['customer_id'], inplace=True)
    return ca_df
```

```python
drop_column(ca_df)
```

```python
ca_df.info()
```

```python
# correcting data types (1)
## ca_df['clv'] = list(map(lambda x: x.str.rstrip('%').astype('float')/100 if (str(x)[-1] == '%') else (x), ca_df['clv']))
ca_df['clv'] = list(map(lambda x: x if (type(x) == float) else float(x[0:-1])/100, ca_df['clv']))
```

```python
# correcting data types (2)
ca_df['number_of_open_complaints'].value_counts(dropna=False)
```

```python
ca_df['number_of_open_complaints'] = list(map(lambda x:x if (type(x) != str) else int(x[2]), ca_df['number_of_open_complaints']))
```

```python
ca_df.info()
```

```python
# standardizing data (1)
ca_df.gender.unique()
```

```python
def clean_gender(x):
    if x in ['M', 'Male']:
        return 'Male'
    elif x in ('F', 'female', 'Femal'):
        return 'Female'
    else:
        return x
```

```python
ca_df['gender'] = list(map(clean_gender, ca_df['gender']))
```

```python
ca_df.gender.unique()
```

```python
# standardizing data (2)
ca_df['state'].value_counts(dropna=False)
```

```python
def clean_state(x):
    if x == 'AZ':
        return 'Arizona'
    elif x == 'Cali':
        return 'California'
    elif x == 'WA':
        return 'Washington'
    else:
        return x
```

```python
ca_df['state'] = list(map(clean_state, ca_df['state']))
```

```python
ca_df['state'].value_counts(dropna=False)
```

```python
# removing duplicates
ca_df.drop_duplicates(inplace=True)
```

```python
ca_df.reset_index(drop=True)
```

```python
ca_df.describe().T
```

```python
# replacing missing values with the mean of the column (DO MEAN FOR NON-ZEROES FROM INCOME!!!!!!!!!!)
ca_df['clv'].fillna(value=ca_df['clv'].mean(axis=0), inplace=True)
ca_df['income'].fillna(value=ca_df['income'].mean(axis=0), inplace=True)
ca_df['monthly_premium_auto'].fillna(value=ca_df['monthly_premium_auto'].mean(axis=0), inplace=True)
ca_df['number_of_open_complaints'].fillna(value=round(ca_df['number_of_open_complaints'].mean(axis=0),0), inplace=True)
ca_df['total_claim_amount'].fillna(value=ca_df['total_claim_amount'].mean(axis=0), inplace=True)
```

```python
ca_df.describe().T
```

# Activity 2

```python
# recategorizing state to zone
def rename_region(x):
    if x == 'California':
        return 'West'
    elif x == 'Washington':
        return 'East'
    elif x == 'Oregon':
        return 'North West'
    elif x == 'Arizona' or 'Nevada':
        return 'Central'
    else:
        return x
```

```python
ca_df['state'] = list(map(rename_region, ca_df['state']))
```

```python
ca_df = ca_df.rename(columns={'state':'zone'})
```

```python
ca_df['zone'].value_counts()
```

```python
ca_df.info()
```

```python
# changing all strings to lower case
ca_df = ca_df.applymap(lambda x:x.lower() if type(x) == str else x)
```
